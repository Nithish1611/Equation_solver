```python
import sympy as sp
import matplotlib.pyplot as plt
def Bisection (f,a,b,Threshold,Max_Steps):
    """ 
    This finction find the root of function f using Bisection method
    
    Input:
    f-->function
    a-->inital value where f(a)>0
    b-->inital value where f(b)<0
    Threshold-->Threshold value if |f(x)|<Threshold function stops and return x as output
    Max_Steps-->Maximum number of steps this function will do after that it will say i cant calculate
    
    Output:
    returns a value c for which |f(c)|< Threshold

    """
    Bdatax=[]
    Bdatay=[]
    n=0
    f=sp.lambdify(x,f)
    while n<Max_Steps:
        c=(a+b)/2
        Bdatax.append(n+1)
        Bdatay.append(f(c))
        if abs(f(c))<Threshold:
            return c,Bdatax,Bdatay
        elif f(c)<0:
            a=c
        else:
            b=c
        n+=1
    if n>=Max_Steps:
        print("Sorry,I cant do it in ",Max_Steps," steps")
        return c,Bdatax,Bdatay
```

```python
def Newton (f,a,Threshold,Max_Steps):
    """ 
    This finction find the root of function f using Newton method
    
    Input:
    f-->function
    a-->inital value
    Threshold-->Threshold value if |f(x)|<Threshold function stops and return x as output
    Max_Steps-->Maximum number of steps this function will do after that it will say i cant calculate
    
    Output:
    returns a value c for which |f(c)|< Threshold

    """
    Ndatax=[]
    Ndatay=[]
    f_dash=sp.diff(f,x)
    f_dash=sp.lambdify(x,f_dash)
    f=sp.lambdify(x,f)
    n=0
    while n<Max_Steps:
        a=a-f(a)/f_dash(a)
        Ndatax.append(n+1)
        Ndatay.append(f(a))
        if abs(f(a))<Threshold:
            return a,Ndatax,Ndatay
        n+=1
    if n>=Max_Steps:
        print("Sorry,I cant do it in ",Max_Steps," steps")
        return a,Ndatax,Ndatay    
    
```

```python
def Secant (f,a,b,Threshold,Max_Steps):
    """ 
    This finction find the root of function f using Secant method
    
    Input:
    f-->function
    a-->inital value from which we start 
    b-->inital value from which we start
    Threshold-->Threshold value if |f(x)|<Threshold function stops and return x as output
    Max_Steps-->Maximum number of steps this function will do after that it will say i cant calculate
    
    Output:
    returns a value c for which |f(c)|< Threshold

    """
    Sdatax=[]
    Sdatay=[]
    n=0
    f=sp.lambdify(x,f)
    while n<Max_Steps:
        c=b-f(b)*((b-a)/(f(b)-f(a)))
        Sdatax.append(n+1)
        Sdatay.append(f(c))
        if abs(f(c))<Threshold:
            return c,Sdatax,Sdatay
        else:
            a=b
            b=c
        n+=1
    if n>=Max_Steps:
        print("Sorry,I cant do it in ",Max_Steps," steps")
        return c,Sdatax,Sdatay
```

```python
x=sp.Symbol('x')
f1=x**3-3.0*x**2-x+9
a=-2
b=0
Threshold=10**-5
Max_Steps=500
N1,Ndatax1,Ndatay1=Newton(f1,a,Threshold,Max_Steps)
B1,Bdatax1,Bdatay1=Bisection(f1,a,b,Threshold,Max_Steps)
S1,Sdatax1,Sdatay1=Secant(f1,a,b,Threshold,Max_Steps)
f2=sp.exp(x)*(x**3-3.0*x**2-x+9)
a=-2
b=0
Threshold=10**-5
Max_Steps=500
N2,Ndatax2,Ndatay2=Newton(f2,a,Threshold,Max_Steps)
B2,Bdatax2,Bdatay2=Bisection(f2,a,b,Threshold,Max_Steps)
S2,Sdatax2,Sdatay2=Secant(f2,a,b,Threshold,Max_Steps)
f3=x**3-2*x+2
a=0
Threshold=10**-5
Max_Steps=500
N3,Ndatax3,Ndatay3=Newton(f3,a,Threshold,Max_Steps)
```

```python
plt.plot(Bdatax1,Bdatay1,label="Bisection Method")
plt.plot(Ndatax1,Ndatay1,label="Newton Method")
plt.xlabel('Number of Steps',fontsize=14)
plt.ylabel('Value of f(x)',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend()
plt.savefig('BNf1.png')
plt.show()
plt.plot(Sdatax1,Sdatay1,label="Secant Method")
plt.xlabel('Number of Steps',fontsize=14)
plt.ylabel('Value of f(x)',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend()
plt.savefig('Sf1.png')
plt.show()
```
